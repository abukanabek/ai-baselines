```python
# https://platform.olimpiada-ai.ro/problems/50

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import json
```

```python
from transformers import pipeline

model_checkpoint = "openai/clip-vit-large-patch14-336"

pipe = pipeline("zero-shot-image-classification", model=model_checkpoint)
```

```python
test_ids = os.listdir('/kaggle/input/the-magic-of-words/test')
```

```python
rows = []

for test_id in tqdm(test_ids):
    with open(os.path.join('/kaggle/input/the-magic-of-words/test', test_id, 'metadata.json'), 'r') as f:
        data_dict = json.load(f)
        
    image_path = os.path.join('/kaggle/input/the-magic-of-words/test', test_id, 'image.png')
    candidate_labels = data_dict['word_choices']

    output = pipe(image_path, candidate_labels=candidate_labels)
    chosen_words = [s['label'] for s in output[:5]]
    row = [test_id] + chosen_words
    rows.append(row)
```

```python
subm = pd.DataFrame(rows, columns=['ID', 'word1', 'word2', 'word3', 'word4', 'word5'])

subm.to_csv("submission.csv", index=False)

subm.head()
```
