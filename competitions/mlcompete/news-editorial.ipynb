```python
# https://platform.olimpiada-ai.ro/problems/67

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
```

```python
train = pd.read_csv("/kaggle/input/news-editorial/train.csv")
test = pd.read_csv("/kaggle/input/news-editorial/test.csv")

train.shape, test.shape
```

```python
train.head(3)
```

```python
train['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=20000, ngram_range=(1, 2))

X, y = train['text'], train['label']
X_test = test['text']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, stratify=y, test_size=0.1, random_state=42)

X_train_tfidf = vectorizer.fit_transform(X_train)
X_valid_tfidf = vectorizer.transform(X_valid)
X_test_tfidf = vectorizer.transform(X_test)

X_train_tfidf.shape, X_valid_tfidf.shape, X_test_tfidf.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_valid_tfidf)

score = accuracy_score(y_valid, y_pred)

print(f'Accuracy: {score:.5f}')
```

```python
y_pred = model.predict(X_test_tfidf)

subm = pd.DataFrame({
    'id': test['id'],
    'label': y_pred
})

subm.to_csv("submission.csv", index=False)
subm.head()
```
